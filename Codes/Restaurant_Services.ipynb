### Adding a New Column in Restaurant Type and Diving the dataset into 4 categories on the basis of service provided by the restaurants.

```python
import pandas as pd
df=pd.read_csv('zomato.csv')
df
```

```python
df_=df[['name','book_table','online_order']].copy()
```

```python
df_['res_ty']=pd.NA
```

```python
df_
```

```python
for i in df_.index:
    if (df_.at[i, 'book_table'] == 'Yes') and (df_.at[i, 'online_order'] == 'Yes'):
        df_.at[i, 'res_ty'] = 'Dine-in / Delivery Online Booking'
        
    elif (df_.at[i, 'book_table'] == 'Yes') and (df_.at[i, 'online_order'] == 'No'):
        df_.at[i, 'res_ty'] = 'Only Dine-in Online Booking'
    
    elif (df_.at[i, 'book_table'] == 'No') and (df_.at[i, 'online_order'] == 'No'):
        df_.at[i, 'res_ty'] = 'Only Dine-in Offline Booking'
    
    elif (df_.at[i, 'book_table'] == 'No') and (df_.at[i, 'online_order'] == 'Yes'):
        df_.at[i, 'res_ty'] = 'Only Online Delivery Booking'
```
