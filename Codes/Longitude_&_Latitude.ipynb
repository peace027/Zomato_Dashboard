```python
df['latitude'] = pd.NA
df['longitude'] = pd.NA
```

```python
from geopy.geocoders import Nominatim
import time

def get_ll(i):
    address = df.at[i, 'address']
    geolocator = Nominatim(user_agent="myGeocoder")
    
    location = None  # Initialize location to None
    
    while True:
        try:
            location = geolocator.geocode(address)

            if location:
                df.at[i, 'latitude'] = location.latitude
                df.at[i, 'longitude'] = location.longitude
                break  # Location found, break the loop
            else:
                # Remove the initial part before the first comma in the address
                comma_index = address.find(',')
                if comma_index != -1:
                    address = address[comma_index + 1:].strip()
                else:
                    break  # If no comma found, break the loop

        except Exception as e:
            # Remove the initial part before the first comma in the address
            comma_index = address.find(',')
            if comma_index != -1:
                address = address[comma_index + 1:].strip()
            else:
                break  # If no comma found, break the loop

    # If location is still not found, set latitude and longitude to None
    if not location:
        df.at[i, 'latitude'] = None
        df.at[i, 'longitude'] = None

for i in df.index:
    if i > 6400:
        get_ll(i)
        print(df.at[i, 'latitude'], df.at[i, 'longitude'], df.at[i, 'address'], i)
        if (i % 100 == 0):
            df.to_csv('modizomato.csv')
            df = pd.read_csv('modizomato.csv')
```
