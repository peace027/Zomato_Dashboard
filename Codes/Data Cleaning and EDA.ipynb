## Zomato Data

- The rate columns comprises of object values like 'New'. [A]
- Missing values in Rate, Phone, Location, Rest_type (Restaurant Type), dish_liked,cuisines, approx_cost. [C]
- approx_cost in object datatype. [V]
- Address given in object time. Need to convert it into interger type in the form of latitude and longitude. [V]
- order_online and book_table are not much efficinet columns, so combining and building a single column is required. [V]
- More than one Cuisines is mentioned in several row. [IC] 
- Menu item columns consists of empty lists. [IC]
- Duplicate and not useful data in dataset. [IC]

```python
#removing NEW from rate.
df.loc[df['rate'] == 'NEW', 'rate'] = '0'
```

```python
#removing - from rate.
df.loc[df['rate'] == '-', 'rate'] = '0'
```

```python
df.groupby('rate').count()
```

```python
#spliting the values in rate.
df['rate'] = df['rate'].str.split('/').str[0]
```

```python
df.groupby('rate').count()
```

```python
df['rate'] = df['rate'].astype(float)
```

```python
df.info()
```

```python
#dealing with all the missing values.
df['phone'] = df['phone'].replace(pd.NA, 'Not Available')
df['rate'] = df['rate'].replace(np.nan, 0.0)
df['location'] = df['location'].replace(pd.NA, 'Not Available')
df['rest_type'] = df['rest_type'].replace(pd.NA, 'Not Available')
df['dish_liked'] = df['dish_liked'].replace(pd.NA, 'Not Available')
df['cuisines'] = df['cuisines'].replace(pd.NA, 'Not Available')
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].replace(pd.NA, 'Not Available')
```

```python
df.isnull().sum()
```

```python
#converting approx_rate datatype
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].replace('Not Available','500')
```

```python
#coerce here mean that any invalid parcing will replace the value to Nan. 
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'].str.replace(',', ''), errors='coerce')


df['approx_cost(for two people)'].fillna(0, inplace=True)

df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(int)
```
